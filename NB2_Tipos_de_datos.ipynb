Python es un lenguaje dinámicamente tipado, esto quiere decir que una misma variable puede tomar valores de distinto _tipo_ en momentos diferentes.

Entonces, al declarar una variable en python asignarás un valor a ésta, el valor asignado tiene un _tipo_. Los tipos de datos primitivos disponibles en python son:

* Integers o enteros.


>```
n = 10
```


* Floats o punto flotante.


>```
pi = 3.14159265
```

* String o cadena.


>```
msj = "Hola mundo!"
```


* Booleans o Booleanos.


>```
flag = True
```

Además de éstos, existen otros tipos de datos que se componen de uno o más datos primitivos.

En python estos son:

* List o listas.
>```
first_five_numbers = [1, 2, 3, 4, 5]
```

* Tuple o tuplas.
>```
pair = (1, 10)
```

* Ranges.
>```
range(5)
```

* Dictionary.
>```
agenda = {'jack': 4098, 'sjoerd': 4127}
```

## Expresiones y declaraciones

Una **expresión** es una combinación de valores, variables y operadores. 

Una **declaración** es una unidad de código que el intérprete de Python puede ejecutar. Hemos visto dos tipos de declaraciones: **impresión** y **asignación**.

````python
millas = 26.2
millas * 1.61
````

La primera línea asigna un valor a millas , pero no tiene ningún efecto visible. La segunda línea es una expresión, por lo que el intérprete la evalúa y muestra el resultado. Entonces aprendemos que una maratón es de unos 42 kilómetros.

   Suma	**+**

### Precedencia de los operadores aritméticos
Al igual que ocurre en matemáticas, en programación también tenemos una **prioridad en los operadores**. Esto significa que si una expresión matemática es precedida por un operador y seguido de otro, el operador más alto en la lista debe ser aplicado por primera vez. Las expresiones con paréntesis se evalúan de dentro a fuera, el paréntesis más interno se evalúa primero. El orden normal de las operaciones es de izquierda a derecha, evaluando en orden los siguientes operadores:

1.   Términos entre paréntesis.
2.   Potenciación y raíces.
3.   Multiplicación y división.
4.   Suma y resta.

En el lenguaje de programación de Python se representan los operadores con el siguiente orden:


1.   ( )
2.   **
3.   *, /, %, //
4.   +, -

Prueba en la consola algunas expresiones matemáticas:

Observa que puedes introducir varias operaciones en una misma línea o expresión. El orden en que se efectúan las operaciones es (en principio) de izquierda a derecha. La expresión 1-2+3, por ejemplo, equivale matemáticamente a ((1-2)+3); por ello la suma y la resta son operadores asociativos por la izquierda.

Si deseamos calcular (1-(2+3)) podemos hacerlo añadiendo paréntesis a la expresión matemática:

````python
(1-(2+3)) 
````

## Funciones de conversión de tipos

Python proporciona funciones integradas que convierten valores de un tipo a otro. La función int toma cualquier valor y lo convierte en un entero, si puede.

````python
int('Hola')
````

**int** puede convertir valores de punto flotante a enteros, pero no se redondea; corta la parte de la fracción:

````python
int(3.99999)
````

````python
int(-2.3)
````

**float** convierte enteros y cadenas en números de coma flotante:

````python
float(32)
````

````python
float('3.14159')
````

Finalmente, **str** convierte su argumento en una cadena:

````python
str(32)
````

````python
str(3.14159)
````

````python
print(70 + float("100.5"))
````

````python
print(str(70) + str(100))
````
